```python
#%% Convert pickle to csv
import pandas as pd
import numpy as np
from tqdm import trange

joints = ['Wrist', 'CMC_thumb', 'MCP_thumb', 'MCP1', 'MCP2', 'MCP3', 'MCP4',
          'IP_thumb', 'PIP1', 'PIP2', 'PIP3', 'PIP4', 'Dip1', 'Dip2', 'Dip3', 'Dip4',
          'Tip_thumb', 'Tip1', 'Tip2', 'Tip3', 'Tip4']

cams = ['cam_0', 'cam_1', 'cam_2', 'cam_3']
postfix = 'DLC_resnet50_Pop_freeReach_0317_mergedApr19shuffle1_190000_full.pickle'
iterables = [[postfix.split('.')[0]], ['right_hand'], joints, ['x', 'y', 'likelihood']]
header = pd.MultiIndex.from_product(iterables, names=['scorer', 'individuals', 'bodyparts', 'coords'])
pickle_paths = [os.path.join(destfolder, cam+postfix) for cam in cams]
paths_to_save = [os.path.join(destfolder, cam+'.csv') for cam in cams]

for pickle_path, path_to_save in zip(pickle_paths, paths_to_save):
    data = pd.read_pickle(pickle_path)
    L = len(data) - 1
    index = np.arange(L)
    df = pd.DataFrame(np.zeros((L, len(joints)*3)), index=index, columns=header)
    for i in trange(L):
        framenum = 'frame' + str(i).zfill(5)
        for k, joint in enumerate(joints):
            coord = data[framenum]['coordinates'][0][k]
            conf = data[framenum]['confidence'][k]
            if len(coord) > 1:
                max_conf_idx = np.argmax(conf)
                # all_points_raw[i, j, k, :] = coord[max_conf_idx]
                # df.iloc[i, 3*k:3*k+2] = coord[max_conf_idx]
                # df.iloc[i, 3*k+2] = conf[max_conf_idx]
                df[postfix.split('.')[0], 'right_hand', joint, 'x'][i] = coord[max_conf_idx][0]
                df[postfix.split('.')[0], 'right_hand', joint, 'y'][i] = coord[max_conf_idx][1]
                df[postfix.split('.')[0], 'right_hand', joint, 'likelihood'][i] = conf[max_conf_idx][0]
                
            elif len(coord) > 0:
                # df.iloc[i ,3*k:3*k+2] = coord
                # df.iloc[i, 3*k+2] = conf
                df[postfix.split('.')[0], 'right_hand', joint, 'x'][i] = coord[0][0]
                df[postfix.split('.')[0], 'right_hand', joint, 'y'][i] = coord[0][1]
                df[postfix.split('.')[0], 'right_hand', joint, 'likelihood'][i] = conf[0]
    
    df.to_csv(path_to_save, mode='w')
```

```python
#%% Good frames
import os
import numpy as np
import pandas as pd

joints = ['Wrist', 'CMC_thumb', 'MCP_thumb', 'MCP1', 'MCP2', 'MCP3', 'MCP4',
          'IP_thumb', 'PIP1', 'PIP2', 'PIP3', 'PIP4', 'Dip1', 'Dip2', 'Dip3', 'Dip4',
          'Tip_thumb', 'Tip1', 'Tip2', 'Tip3', 'Tip4']

scheme = [
  ["Wrist", "CMC_thumb"],
  ["CMC_thumb", "MCP_thumb"],
  ["MCP_thumb", "IP_thumb"],
  ["IP_thumb", "Tip_thumb"],
  ["Wrist", "MCP1"],
  ["Wrist", "MCP2"],
  ["Wrist", "MCP3"],
  ["Wrist", "MCP4"],
  ["MCP1", "PIP1"],
  ["MCP2", "PIP2"],
  ["MCP3", "PIP3"],
  ["MCP4", "PIP4"],
  ["PIP1", "Dip1"],
  ["PIP2", "Dip2"],
  ["PIP3", "Dip3"],
  ["PIP4", "Dip4"],
  ["Dip1", "Tip1"],
  ["Dip2", "Tip2"],
  ["Dip3", "Tip3"],
  ["Dip4", "Tip4"]]

no_wrist_scheme = [
  ["CMC_thumb", "MCP_thumb"],
  ["MCP_thumb", "IP_thumb"],
  ["IP_thumb", "Tip_thumb"],
  ["MCP1", "PIP1"],
  ["MCP2", "PIP2"],
  ["MCP3", "PIP3"],
  ["MCP4", "PIP4"],
  ["PIP1", "Dip1"],
  ["PIP2", "Dip2"],
  ["PIP3", "Dip3"],
  ["PIP4", "Dip4"],
  ["Dip1", "Tip1"],
  ["Dip2", "Tip2"],
  ["Dip3", "Tip3"],
  ["Dip4", "Tip4"]]

wrist_scheme = [
  ["Wrist", "CMC_thumb"],
  ["CMC_thumb", "MCP_thumb"],
  ["MCP_thumb", "IP_thumb"],
  ["IP_thumb", "Tip_thumb"],
  ["Wrist", "MCP1"],
  ["Wrist", "MCP2"],
  ["Wrist", "MCP3"],
  ["Wrist", "MCP4"]]

dist_meds_real = {'Wrist-CMC_thumb': [23.77, 20.83, 23.26],
 'CMC_thumb-MCP_thumb': [18.73, 24.96, 22.56],
 'MCP_thumb-IP_thumb': [14.25, 18.23, 17.72],
 'IP_thumb-Tip_thumb': [11.66, 9.87, 10.53],
 'Wrist-MCP1': [43.86, 42.38, 35.63],
 'Wrist-MCP2': [42.81, 39.57, 35.35],
 'Wrist-MCP3': [42.81, 37.46, 36.32],
 'Wrist-MCP4': [40.50, 40.22, 36.89],
 'MCP1-PIP1': [26.32, 27.40, 25.19],
 'MCP2-PIP2': [30.07, 33.44, 30.00],
 'MCP3-PIP3': [30.27, 32.70, 28.53],
 'MCP4-PIP4': [21.95, 24.52, 24.44],
 'PIP1-Dip1': [18.01, 16.92, 13.58],
 'PIP2-Dip2': [23.08, 20.38, 18.19],
 'PIP3-Dip3': [21.98, 21.16, 20.16],
 'PIP4-Dip4': [18.29, 18.25, 14.78],
 'Dip1-Tip1': [11.54, 12.82, 11.55],
 'Dip2-Tip2': [16.18, 15.39, 12.98],
 'Dip3-Tip3': [14.72, 13.34, 14.15],
 'Dip4-Tip4': [11.17, 11.64, 10.57]}

df = pd.read_csv(os.path.join(pickle_folder, 'output_3d_data_lpf_cut.csv'))
# df = df_filt.copy()
connection_dict = {}
arg_accurate = {}
dist_meds = {}
dist_means = {}
dist_stds = {}
dist_outliers = {}
for connection in scheme:
    poi1 = connection[0]
    poi2 = connection[1]
    poi1_coord = np.stack([df[poi1+'_x'], df[poi1+'_y'], df[poi1+'_z']], axis=1)
    poi2_coord = np.stack([df[poi2+'_x'], df[poi2+'_y'], df[poi2+'_z']], axis=1)
    dist = np.linalg.norm(poi1_coord - poi2_coord, axis=1)
    connection_dict[poi1+'-'+poi2] = dist
    dist_med = np.median(dist[np.isfinite(dist)])
    dist_std = np.std(dist[np.isfinite(dist)])
    dist_mean = np.mean(dist[np.isfinite(dist)])
    dist_meds[poi1+'-'+poi2] = dist_med
    dist_stds[poi1+'-'+poi2] = dist_std
    dist_means[poi1+'-'+poi2] = dist_mean
    threshold = dist_meds_real[poi1+'-'+poi2][-1]
    # threshold = dist_med
    dist[np.isnan(dist)] = -100
    # arg_accurate[poi1+'-'+poi2] = np.argwhere((dist > (threshold - threshold*0.2)) & 
    #                                           (dist < (threshold + threshold*0.2)))
    # arg_accurate[poi1+'-'+poi2] = np.argwhere((dist > (threshold - dist_std)) & 
    #                                           (dist < (threshold + dist_std)))
    arg_accurate[poi1+'-'+poi2] = np.argwhere((dist > (threshold*0.9)) & 
                                              (dist < (threshold*1.1)))

good_reconstruction_indices = []
good_number = []
where_good = {}
for i in range(len(df)):
    idx_threshold = 0
    where_good[i] = []
    for connection in scheme:
        poi1 = connection[0]
        poi2 = connection[1]
        if i in arg_accurate[poi1+'-'+poi2]:
            idx_threshold += 1
            where_good[i].append(poi1)
            where_good[i].append(poi2)
    
    good_number.append(idx_threshold)
    if idx_threshold >= len(scheme)-10:
        good_reconstruction_indices.append(i)
    
    where_good[i] = set(where_good[i])

where_good_joint = {}
where_joint = {}
chose_good_joint = {}

for joint in joints:
    where_good_joint[joint] = []
    where_joint[joint] = []
    chose_good_joint[joint] = []

for i in range(len(df)):
    for joint in joints:
        if joint in where_good[i]:
            where_joint[joint].append(i)
            

for idx in good_reconstruction_indices:
    for joint in joints:
        if joint in where_good[idx]:
            where_good_joint[joint].append(idx)

N = 1000
chosen_indices = np.random.choice(good_reconstruction_indices, N)
for idx in chosen_indices:
    for joint in joints:
        if joint in where_good[idx]:
            chose_good_joint[joint].append(idx)
    
print('Mean of standard deviation of inferred connection lengths is: {}'.format(np.mean(list(dist_stds.values()))))
print('Length of good reconstructions indices: {}'.format(len(good_reconstruction_indices)))
```

```python
#%% Extract specific frames
import os
import re
import cv2
import numpy as np
from tqdm import tqdm

def extract_specific_frames(vidpath, frame_counts, path_to_save): 
    if not os.path.exists(vidpath):
        print('Video does not exist.')
        return
    
    if not os.path.exists(path_to_save):
        print(path_to_save + ' does not exsit.')
        folder_input = input('Do you want to create this path (folder)? (y/n) ')
        if folder_input is 'y':
            os.mkdir(path_to_save)
        elif folder_input is 'n':
            return
        else:
            print('Wrong input.')
            return
            
    cap = cv2.VideoCapture(vidpath) 
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_num = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = frame_num/fps
    count = len(frame_counts)
        
    with tqdm(total=count) as pbar:
        for frame_count in frame_counts:
            cap.set(cv2.CAP_PROP_POS_FRAMES, frame_count)
            ret, frame = cap.read()
            cv2.imwrite(os.path.join(path_to_save, 'img' + str(frame_count).zfill(6) + '.png'), frame)
            pbar.update(1)
            
    print('\n{} frames were extracted.'.format(count))
    
vidfolder = os.path.join(os_path, session_folder)
vidnames = [cam+'.avi' for cam in cams]
vidpaths = [os.path.join(vidfolder, vidname) for vidname in vidnames]

output_folder = '/home/myp7435/Pop_freeReach_0317_merged-Min-2020-04-19/labeled-data/'
mvb_iter = 2
folders = [cam+'_'+os.path.basename(vidfolder)+'_b'+str(mvb_iter) for cam in cams]
paths_to_save = [os.path.join(output_folder, folder) for folder in folders]

image_indices = chosen_indices

for vidpath, path_to_save in zip(vidpaths, paths_to_save):
    extract_specific_frames(vidpath, image_indices, path_to_save)
```

```python
#%% Multi-view Bootstrapping
import os
import cv2
import numpy as np
import pandas as pd
from numpy import array as arr
from utils.calibration_utils import load_intrinsics, load_extrinsics


def get_rtvec(M):
    rvec = cv2.Rodrigues(M[:3, :3])[0].flatten()
    tvec = M[:3, 3].flatten()
    return rvec, tvec

vid_indices = ['0', '1', '2', '3']
intrinsics = load_intrinsics(calib_path, vid_indices)
extrinsics = load_extrinsics(calib_path)

df = pd.read_csv(os.path.join(pickle_folder,'output_3d_data_lpf_cut.csv'))
data_3d = df.copy()

L = len(image_indices)
joints = ['Wrist', 'CMC_thumb', 'MCP_thumb', 'MCP1', 'MCP2', 'MCP3', 'MCP4',
          'IP_thumb', 'PIP1', 'PIP2', 'PIP3', 'PIP4', 'Dip1', 'Dip2', 'Dip3', 'Dip4',
          'Tip_thumb', 'Tip1', 'Tip2', 'Tip3', 'Tip4']
# joints = ['pointX','pointY','pointZ']

iterables = [['Min'], ['right_hand'], joints, ['x', 'y']]
header = pd.MultiIndex.from_product(iterables, names=['scorer', 'individuals', 'bodyparts', 'coords'])

data_2ds = []

for vid_idx, path_to_save in zip(vid_indices, paths_to_save):
    cameraMatrix = np.matrix(intrinsics[vid_idx]['camera_mat'])
    distCoeffs = np.array(intrinsics[vid_idx]['dist_coeff'])
    Rt = np.matrix(extrinsics[vid_idx])
    rvec, tvec = get_rtvec(Rt)
    
    out = os.path.basename(path_to_save)
    index = [os.path.join('labeled-data', out, 'img' + str(frame_count).zfill(6) + '.png') 
              for frame_count in image_indices]
    
    df_2d = pd.DataFrame(np.zeros((L, len(joints)*2)), index=index, columns=header)
    
    for i, bp in enumerate(joints):
        x = data_3d[bp+'_x'][image_indices].copy()
        y = data_3d[bp+'_y'][image_indices].copy()
        z = data_3d[bp+'_z'][image_indices].copy()
        
        objectPoints = np.vstack([x,y,z]).T
        coord_2d = np.squeeze(cv2.projectPoints(objectPoints, rvec, tvec, cameraMatrix, distCoeffs)[0], axis=1)
        where_negative = np.argwhere(coord_2d[:,0] < 0)
        where_out1 = np.argwhere(coord_2d[:,0] >= 2048)
        where_out2 = np.argwhere(coord_2d[:,1] >= 1536)
        coord_2d[where_negative, :] = np.nan
        coord_2d[where_out1, :] = np.nan
        coord_2d[where_out2, :] = np.nan
        
        df_2d.iloc[:, 2*i:2*i+2] = coord_2d
    
    if not os.path.exists(path_to_save):
        os.mkdir(path_to_save)
    df_2d.to_hdf(os.path.join(path_to_save,'CollectedData_Min.h5'), key='df_with_missing', mode='w')
    data_2ds.append(df_2d)
```
